```python
import pandas as pd
```

```python
covid_jabar = pd.read_csv('Data COVID-19 Jawa Barat.csv')
covid_jabar.info()
```

```python
covid_jabar
```

```python
covid_jabar_cfm = covid_jabar[['tanggal','nama_kab','confirmation','confirmation_selesai','confirmation_meninggal']]
covid_jabar_cfm
```

```python
covid_jabar_cfm['tanggal'] = pd.to_datetime(covid_jabar_cfm['tanggal'])
covid_jabar_cfm.info()
```

```python
covid_jabar_cfm_240422 = covid_jabar_cfm.loc[(covid_jabar_cfm['tanggal'] == '2022-04-24')].reset_index(drop=True)
covid_jabar_cfm_240422.info()
covid_jabar_cfm_240422
```

```python
covid_jabar_cfm_240422 = covid_jabar_cfm_240422.drop(columns=['tanggal'])
covid_jabar_cfm_240422
```

```python
covid_jabar_cfm_240422 = covid_jabar_cfm_240422.drop([12]).reset_index(drop=True)
covid_jabar_cfm_240422
```

```python
covid_jabar_final = pd.DataFrame(covid_jabar_cfm_240422)
covid_jabar_final['confirmation_perawatan'] = covid_jabar_final['confirmation'] - (covid_jabar_final['confirmation_selesai'] + covid_jabar_final['confirmation_meninggal'])
covid_jabar_final
```

```python
covid_jabar_final = covid_jabar_final[['nama_kab','confirmation','confirmation_perawatan','confirmation_selesai','confirmation_meninggal']]
covid_jabar_final
```

```python
covid_jabar_final.info()
```

```python
covid_jabar_final.to_csv('Covid19_Jabar.csv')
```
